```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
gas=pd.read_csv('./Matplotlib Tutorial/gas_prices.csv')
gas.head(10)
```

```python
plt.figure(figsize=(8,5))
plt.title('Gas price hike over year in different country')
plt.xlabel('Years')
plt.ylabel('Price in USD($)')
plt.xticks(gas.Year[::3]) #::3 means x axis in interval of 3
plt.plot(gas.Year,gas.USA,'b.-',label='USA') #(.- gives the . at every interval)
plt.plot(gas.Year,gas.Canada, 'r.-',label='Canada')
plt.plot(gas.Year,gas['South Korea'],'g.-',label='South Korea')
plt.legend()
plt.plot()
```

```python
#Other way
plt.figure(figsize=(8,5))
plt.title('Gas price hike over year in different country',fontdict={'fontweight':'bold','fontsize':18})
plt.xlabel('Years')
plt.ylabel('Price in USD($)')
plt.xticks(gas.Year[::3]) #::3 means x axis in interval of 3
#to have 2011 at the end
#plt.xticks(gas.Year[::3].tolist()+[2011])
for country in gas:
    if country!='Year':
        print(country)
        plt.plot(gas.Year,gas[country], marker='.')
plt.plot()
```

```python
#FIFA DATA
fifa=pd.read_csv('./Matplotlib Tutorial/fifa_data.csv')
fifa.head()
```

```python
bins=[40,50,60,70,80,90,100]
plt.hist(fifa.Overall,bins=bins,color='#b33c00')
plt.title('Distribution of Player Skills in FIFA',fontdict={'fontstyle':'oblique','fontweight':'bold','fontsize':16})
plt.xlabel('Skill Level')
plt.ylabel('No of Players')
plt.xticks(bins)
plt.plot()
```

```python
#pie chart
left=fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
right=fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]
right
#left
```

```python
plt.pie([left,right],labels=['Left Foot','Right Foot'],colors=['g','r'],autopct='%.2f %%') #autopct  is used to denote the %
plt.title('Comparison Of Foot between FIFA Players')
plt.plot()
```

```python
fifa.Weight =[int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]
fifa.Weight
```

```python
light=fifa.loc[fifa.Weight<125].count()[0]
light_medium=fifa.loc[(fifa.Weight>=125)&(fifa.Weight<150)].count()[0]
medium=fifa.loc[(fifa.Weight>=150)&(fifa.Weight<175)].count()[0]
medium_heavy=fifa.loc[(fifa.Weight>=175)&(fifa.Weight<200)].count()[0]
heavy=fifa.loc[(fifa.Weight>=200)].count()[0]
```

```python
labels=['under 125','125-150','150-175','175-200', 'Above 200']
plt.title("Comparison of Player Weight in lps",fontdict={'fontstyle':'oblique','fontweight':'bold','fontsize':16})
plt.pie([light,light_medium,medium,medium_heavy,heavy],labels=labels,autopct="%.2f %%")
plt.plot()
```

```python
# In above Graph we can under 125 and above 200 are meshed up to avaoid this
labels=['under 125','125-150','150-175','175-200', 'Above 200']
plt.title("Comparison of Player Weight in lps",fontdict={'fontstyle':'oblique','fontweight':'bold','fontsize':16})
#explode is used to break the chart. Here we can break 125-150, under 125 and above200
explode=[0.4,0.2,0,0,0.4]
plt.pie([light,light_medium,medium,medium_heavy,heavy],labels=labels,autopct="%.2f %%",pctdistance=0.8,explode=explode)
plt.plot()
```

```python
#Box and Whisker Plot
plt.figure(figsize=(6,10))
Barcelona=fifa.loc[fifa.Club=='FC Barcelona']['Overall']
ManUnited=fifa.loc[fifa.Club=='Manchester United']['Overall']
ManCity=fifa.loc[fifa.Club=='Manchester City']['Overall']
Paris=fifa.loc[fifa.Club=='Paris Saint-Germain']['Overall']
labels=['Barcelona', 'Man Uni', 'Manchester City', 'Paris']
#plt.boxplot([Barcelona, ManUnited, ManCity, Paris],labels=labels)
plt.xlabel('City')
plt.ylabel('FIFA Overall rating')
#For styling
boxes=plt.boxplot([Barcelona, ManUnited, ManCity, Paris],labels=labels, patch_artist=True, medianprops={'linewidth':2})
for  box in boxes['boxes']:
    #set edge color
    box.set(color='b', linewidth= 2)
    #fill color
    box.set(facecolor='r')
plt.show()

#here patch_artist is used to fill the color in the box, medainprops is to inc the style of the line inside the box
```
