```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=[1,2,3,4]
y=[2,4,6,8]
plt.plot(x,y)
plt.title('My First Graph!', fontdict={'fontname': 'Times New Roman', 'fontsize': '18'}) # fontdict is to add details to text
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
#to avoid decimal values we can ues 'ticks'
plt.xticks(x)
plt.yticks(y)
plt.show()
```

```python
x=[1,2,3,4]
y=[2,4,6,8]
plt.plot(x,y, label='2 Times of X', color='red', linestyle='--', linewidth='3', marker='*',markersize='5', markeredgecolor='g')
plt.title('My First Graph!', fontdict={'fontname': 'Times New Roman', 'fontsize': '18'}) # fontdict is to add details to text
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
#to avoid decimal values we can ues 'ticks'
plt.xticks(x)
plt.yticks(y)
plt.legend()
plt.show()
```

```python
#when we add one more axes
x=[1,2,3,4]
y=[2,4,6,8]
x2=np.arange(0,4.5,0.5)
print(x2)
plt.plot(x2, x2**2, 'g', label='x^2')
plt.plot(x,y, label='2 Times of X')
plt.title('My First Graph!', fontdict={'fontname': 'Times New Roman', 'fontsize': '18'}) # fontdict is to add details to text
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
#to avoid decimal values we can ues 'ticks'
plt.xticks(x)
plt.yticks(y)
plt.legend()
plt.show()
```

```python
#To make a projection
x=[1,2,3,4]
y=[2,4,6,8]
#Resize Graph
plt.figure(figsize=(5,3),dpi=100)  #dpi stands for resolution
x2=np.arange(0,4.5,0.5)
print(x2)
#plt as line
plt.plot(x2[:7], x2[:7]**2, 'g', label='x^2')
#plt as --
plt.plot(x2[6:], x2[6:]**2, 'g--', label='x^2')
plt.plot(x,y, label='2 Times of X')
plt.title('My First Graph!', fontdict={'fontname': 'Times New Roman', 'fontsize': '18'}) # fontdict is to add details to text
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
#to avoid decimal values we can ues 'ticks'
plt.xticks(x)
plt.yticks(y)
plt.legend()
plt.savefig('My Graph.png',dpi=300) #To save the Grap
plt.show()
```
